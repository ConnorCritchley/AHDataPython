```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../../Data/grades/sat_gpa_v1.csv")
df.info()
```

```python
#Creating a split

# separate features and target
X = df.drop("fy_gpa", axis=1)
y = df["sat_sum"]

# split data into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=12)
```

```python
#Fitting
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
```

```python
#Evaluate
pred = regr.predict(X_test)
# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, pred))
```

```python
sns.lmplot(df, x='sat_sum', y='fy_gpa')
plt.title("Linear Regreasson of sat_sum to fy_gpa")
```

### Conclusion
#### Seeing as how the linear regression model was able to completely predict the data, and the graph displays a line with a strong slope, I would say the sat score is strongly positively related to first year gpa.
